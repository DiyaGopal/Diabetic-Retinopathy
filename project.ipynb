```python

# Step 1: Import Libraries
import numpy as np
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
# Step 2: Load and Preprocess Dataset

import os
import pandas as pd
import cv2
import numpy as np

IMG_SIZE = 128

# Paths
images_path = r'C:\Users\DIYA GOPAL\OneDrive\Desktop\DIP\data\colored_images'
csv_path = r'C:\Users\DIYA GOPAL\OneDrive\Desktop\DIP\data\train.csv'

# Read CSV
df = pd.read_csv(csv_path)
print(df.head())

# Function to load image by searching all subfolders
def load_image(img_name):
    for root, dirs, files in os.walk(images_path):
        for file in files:
            if img_name in file and file.endswith('.png'):
                img_path = os.path.join(root, file)
                img = cv2.imread(img_path)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
                return img
    return None

# Load dataset
data = []
labels = []

for idx, row in df.iterrows():
    img_name = row['id_code']
    label = row['diagnosis']

    img = load_image(img_name)
    if img is not None:
        data.append(img)
        labels.append(label)
    else:
        print(f"Image not found for {img_name}")

# Convert to numpy arrays
data = np.array(data) / 255.0
labels = np.array(labels)

print(f"Loaded {len(data)} images and labels")
```

```python
# Step 3: Encode Labels
lb = LabelBinarizer()
labels = lb.fit_transform(labels)
```

```python
# Step 4: Split Data
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42, stratify=labels)
```

```python
# Step 5: Build CNN Model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(lb.classes_), activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Step 6: Train Model
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```python
# Step 7: Evaluate Model
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training vs Validation Accuracy')
plt.show()

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Training vs Validation Loss')
plt.show()

```

```python
# Step 8: Save the model
model.save('eye_disease_detector.h5')
print("Model saved as eye_disease_detector.h5")
```

```python
# Step 9: REAL-TIME PREDICTION from TEST DATA

# Predict on Test Set
predictions = model.predict(X_test)
predicted_labels = lb.inverse_transform(predictions)
true_labels = lb.inverse_transform(y_test)

# Function to display image with true and predicted label
def show_prediction(index):
    plt.imshow(X_test[index])
    plt.axis('off')
    plt.title(f"True: {true_labels[index]}\nPredicted: {predicted_labels[index]}")
    plt.show()

# Show 5 random predictions
import random
for _ in range(5):  # Show 5 random test samples
    idx = random.randint(0, len(X_test)-1)
    show_prediction(idx)
```
